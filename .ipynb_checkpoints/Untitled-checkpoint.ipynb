```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statsmodels.formula.api import ols
import statsmodels.api as sm
import re
plt.rcParams['agg.path.chunksize'] = 10000
pd.set_option('display.max_columns', None)
```

```python
train_df = pd.read_csv("train.csv", sep = ',', encoding = 'unicode_escape')
```

```python
test_df = pd.read_csv("test.csv", sep = ',', encoding = 'unicode_escape')
```

```python
train_df
```

```python
train_df["grade"] = train_df["grade"].map({'A':7, 'B':6, 'C':5, 'D':4, 'E':3, 'F':2, 'G':1})
```

```python
train_df["term"] = train_df["term"].replace([' 36 months'], 36)
train_df["term"] = train_df["term"].replace([' 60 months'], 60)
```

```python
train_df = train_df.sort_values(by=['int_rate'])
```

```python
train_df
```

```python
test_df
```

```python
train_df = train_df.dropna()
```

```python
train_df['emp_length'].isna().sum()
```

```python
train_df["emp_length"] = train_df.emp_length.str.extract('(\d+)')
```

```python
train_df = train_df.drop(['grade'], axis=1)
```

```python
train_df = train_df.drop(['funded_amnt'], axis=1)
train_df = train_df.drop(['funded_amnt_inv'], axis=1)
train_df = train_df.drop(['installment'], axis=1)
train_df = train_df.drop(['id'], axis=1)
```
